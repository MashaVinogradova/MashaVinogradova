```python
import pandas as pd
from datetime import datetime

users = pd.read_excel('users_hw.xlsx') # данные по клиентам
payments = pd.read_excel('payments_hw.xlsx') # данные по платежам
workouts = pd.read_csv('workouts_hw.csv') # данные по тренировкам

pd.set_option('display.float_format', lambda x: '%.2f' % x) # удобно выводить большие числа
```

```python
payments.info() # смотрим датафрейм payments
```

```python
payments['payment_date'] = pd.to_datetime(payments['payment_date'], infer_datetime_format=True) # даты в формат datetime
payments.info()
```

```python
payments
```

```python
payments_2020 = payments.loc[payments['payment_date'] >= '2020-01-01'] # выбираем платежи только за 2020 год
payments_2020
```

```python
payments_2020['currency'].describe()
```

```python
USD_RATE = 73.14 # курс доллара

def map_currency(currency):
    if currency == 'RUR':
        return 1
    else: 
        return USD_RATE 
    # если это USD, то берем значение переменной USD_RATE, если RUR — ставим значение 1

    
```

```python
payments_2020['fx_rate'] = payments_2020['currency'].apply(map_currency) # определяем функцию для заполнения столбца fx_rate 
payments_2020
```

```python
payments_2020['amount_rub'] = payments_2020['amount'] * payments_2020['fx_rate'] # еще один столбец amount_rub с суммой платежа в рублях.
payments_2020
```

```python
payments_2020['workout_type'].describe()
```

```python
revenue = payments_2020['amount'].groupby([payments_2020['payment_date'].dt.month, payments_2020['workout_type']]).sum() # датафрейм с разбивкой выручки по месяцам и по типам тренировок
revenue
```

```python
workouts # смотрим датафрейм `workouts`
```

```python
paid_workouts = workouts.loc[workouts['trainer_department'] != 'Sales'] # исключаем из датафрейма тренеров из департамента `Sales'
paid_workouts
```

```python
workouts_by_trainer = paid_workouts.groupby('trainer_id')['workout_id'].nunique() # делаем группировку по `trainer_id` и выведи количество уникальных тренировок по `workout_id`

pd.DataFrame(workouts_by_trainer)

workouts_by_trainer_df = workouts_by_trainer.reset_index() 

workouts_by_trainer_df
```

```python
max_trainings = workouts_by_trainer_df['workout_id'].max() # максимальное количество тренировок
max_trainings
```

```python
workouts_by_trainer_df.loc[workouts_by_trainer_df['workout_id'] == max_trainings] # id тренера, у которого количество тренировок равно значению `max_trainings`
```

```python
users['free_workout_datetime'] = pd.to_datetime(users['free_workout_datetime'])   # даты в формат datetime

november_users = users.loc[(users['free_workout_datetime']>=datetime(2020,11,1))   # отобрали клиентов, кто начал заниматься с 1 ноября
                           &(users['free_workout_datetime']<datetime(2020,12,1))]  # по 30 ноября 2020 года

nov_users_payments = pd.merge(payments_2020, november_users) # объединение получившегося набора данных с платежами из датафрейма payments по user_id

nov_users_payments.groupby('workout_type')['amount_rub'].sum()
```

```python
user_regions = users[['user_id','region']] # новый датафрейм user_regions из колонок user_id и region датафрейма users
user_regions
```

```python
USD_RATE = 73.14

def map_currency(currency):
    if currency == 'RUR':
        return 1
    else: 
        return USD_RATE 
    
payments['fx_rate'] = payments['currency'].apply(map_currency)

payments['amount_rub'] = payments['amount'] * payments['fx_rate']

payments
```

```python
user_payments = payments[['user_id','amount_rub']] # новый датафрейм user_payments из колонок user_id и amount_rub датафрейма payments
user_payments
```

```python
# объединение датафреймов user_regions и user_payments
regions_payments = pd.merge(user_regions, user_payments)

revenue_by_region = regions_payments.dropna().groupby('region')['amount_rub'].sum() # группировка данных о выручке по региону

revenue_by_region_df = pd.DataFrame(revenue_by_region) # перевела в Data Frame

revenue_by_region_df_in = revenue_by_region_df.reset_index() # добавила индекс
```

```python
revenue_by_region_df_in.sort_values(by='amount_rub').head() # вывод на экран топ-5 регионов с самой маленькой выручкой.
 
```

```python
revenue_by_region_df_in.sort_values(by='amount_rub', ascending=False).head() # вывод на экран топ-5 регионов с самой большой выручкой
```

```python
revenue_by_region_df_in.sort_values(by='amount_rub', ascending=False).to_excel('task_4.xlsx') # сохранение в Excel
```
